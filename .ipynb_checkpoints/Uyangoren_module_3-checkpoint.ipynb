### PROBLEM 1
**Fourier transform of a simple function**

1. Calculate the coefficients in the discrete Fourier transforms of the modulated sine wave $y_n = \sin(\pi n/N) sin(20\pi n/N)$ sampled at $N = 1000$ evenly spaced points using your own user-defined function. *(15 pts.)*

2. Make a plot of their amplitudes similar to the plot shown in Fig. 7.4 of the Newman text. *(15 pts.)*


````python
import numpy as np
from numpy import zeros
import cmath
from cmath import exp, pi
import matplotlib.pyplot as plt

#define the modulated sine wave
def y(n):
  y = np.sin(pi * n / N)*np.sin(20* pi * n/N)
  return y

#function for determining the discrete Fourier transforms of y
def dft(y):
  c = np.zeros(N//2 +1, complex)     #make an array of complex numbers with zero values to be used in storing the coefficients, length is N//2
  for  k in range(N//2 +1):           #determine DFT oefficients of y using the equation for C_n for different values of n and store in the array
    for n in range(N):
      c[k] += y[n]*exp(-2j*pi*k*n/N)    #C_n formula for different values of n
  return c

N = 1000


x_ax = np.linspace(0,N,N)  #create array containing the x-axis index values (0 to 1000) for plotting
y_ax = y(x_ax)             #create array containing the y-axis (DFT of values) for plotting.
coeff = dft(y_ax)          #store DFT coefficients into this array

#plot original sine wave signal
plt.plot(y_ax)
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Original Signal")
plt.show()

#plot coefficients of DFT
plt.plot(abs(coeff))
plt.xlabel("Frequency")
plt.ylabel("Amplitude")
plt.title("Magnitude of DFT Components")
plt.show()



````

Now you can start to write code for this problem!

````python
import matplotlib.pyplot as plt
import numpy as np

coeff = np.fft.rfft(dow)      #determine the DFT coefficients of dow

for i in range(len(coeff)):
  if i > 0.1*len(coeff):      #remove the higher frequency values (>10%) in order to eliminate noise signals/irrelevant signals (Irrelevant because they change more rapidly and might be less relevant for analysis)
    coeff[i] = 0
  else:
    coeff[i] = coeff[i]

inverse = np.fft.irfft(coeff)   #perform inverse transform to reconstruct original signal, i.e. Dow Data

#plot original signal(blue) and reconstructed signal (red)
plt.figure(figsize=(11, 8))
plt.plot(dow, color="blue")
plt.plot(inverse, color="red")
plt.xlabel("Index")
plt.ylabel("DOW")
plt.title("Daily closing value of the Dow Jones Industrial Average (2006-2010)")
plt.show()
````

In the code, we were able to remove the components that rapidly change, i.e., those with higher frequencies. These were removed because they change so rapidly that, in analysis of the data, they may be considered irrelevant or that they are only short-term fluctuations due to random market activities on a day-to-day basis.

Upon reconstructing the original signal, we only obtained the significant and larger fluctuations in the Dow Jones Industrial Average. Hence, from this plot, we see significant rise and drops in the Dow. These changes could be attributed to political and economic events during the given timeline.
